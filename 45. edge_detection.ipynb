```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('sammy_face.jpg')
plt.imshow(img)
```

```python
edges = cv2.Canny(image=img,threshold1=127,threshold2=127)
plt.imshow(edges)
```

```python
edges = cv2.Canny(image=img,threshold1=0,threshold2=255)
plt.imshow(edges)
```

```python
median_val = np.median(img)
median_val
```

```python
# LOWER THRESHOLD TO EITHER 0 OR 70% OF MEDIAN VALUE, WHICHEVER IS GREATER
lower = int(max(0,0.7*median_val))
# UPPER THRESHOLD TO EITHER 255 OR 130% OF MEDIAN VALUE, WHICHEVER IS SMALLER
upper = int(min(255,1.3*median_val))
```

```python
edges = cv2.Canny(image=img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
plt.imshow(blurred_img)
```

```python
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(7,7))
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize=(5,5))
edges = cv2.Canny(image=blurred_img,threshold1=lower,threshold2=upper+50)
plt.imshow(edges)
```
