```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog_backpack.jpg')
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
new_img = img_rgb.copy()
```

```python
new_img = cv2.flip(new_img,0)
plt.imshow(new_img)
```

```python
pt1 = (200,380)
pt2 = (600,700)
```

```python
cv2.rectangle(img_rgb,pt1=pt1,pt2=pt2,color=(255,0,0),thickness=10)
plt.imshow(img_rgb)
```

```python
vertices = np.array([ [250,700],[425,400],[600,700] ],np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape(-1,1,2)
```

```python
pts.shape
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.polylines(img_rgb,[pts],isClosed=True,color=(0,0,255),thickness=20)
plt.imshow(img_rgb)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.fillPoly(img_rgb,[pts],color=(0,0,255))
plt.imshow(img_rgb)
```
