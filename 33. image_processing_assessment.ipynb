```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def display(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display(img)
```

```python
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY) # with all 3 color channels
display(thresh1)
```

```python
img = cv2.imread('giraffes.jpg',0) # gray scale image
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
display(thresh1,cmap='gray')
```

```python
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display(img)
```

```python
kernel = np.ones(shape=(4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
img = cv2.imread('giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
result = cv2.filter2D(img,-1,kernel)
display(result)
```

```python
img = cv2.imread('giraffes.jpg',0)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
display(sobelx,cmap='gray')
```

```python
img = cv2.imread('giraffes.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)

plt.title('HISTOGRAM FOR GIRAFFES')
```
