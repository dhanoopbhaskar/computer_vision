```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('00-puppy.jpg')
```

```python
type(img) # if the type is NoneType the read operation failed
```

```python
img.shape
```

```python
# MATPLOTLIB --> RGB
# OPENCV --> BGR
plt.imshow(img)
```

```python
fix_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
img_gray = cv2.imread('00-puppy.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray,cmap='gray')
```

```python
new_img = cv2.resize(fix_img,(1000,400))
```

```python
plt.imshow(new_img)
```

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(fix_img,(0,0),fix_img,w_ratio,h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img,0)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,1)
plt.imshow(new_img)
```

```python
new_img = cv2.flip(fix_img,-1)
plt.imshow(new_img)
```

```python
cv2.imwrite('new_img.jpg', fix_img)
```
