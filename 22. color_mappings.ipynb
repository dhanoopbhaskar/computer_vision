```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('00-puppy.jpg')
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
```

```python
img_hls = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(img_hls)
```
